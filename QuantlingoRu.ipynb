```python
from transformers import MarianMTModel, MarianTokenizer
from optimum.onnxruntime import ORTModelForSeq2SeqLM
import torch

# Конфигурация
MODEL_NAME_RU_EN = "Helsinki-NLP/opus-mt-ru-en"
MODEL_NAME_EN_RU = "Helsinki-NLP/opus-mt-en-ru"
ONNX_DIR = "./onnx_models"

# Загрузка модели и токенизатора
def export_model(model_name, direction):
    # Конвертация в ONNX через Optimum
    model = ORTModelForSeq2SeqLM.from_pretrained(
        model_name,
        export=True,
        provider="CPUExecutionProvider"
    )

    # Сохранение ONNX модели и токенизатора
    model.save_pretrained(f"{ONNX_DIR}/{direction}")
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    tokenizer.save_pretrained(f"{ONNX_DIR}/{direction}")

# Экспорт обеих моделей
export_model(MODEL_NAME_RU_EN, "ru-en")
export_model(MODEL_NAME_EN_RU, "en-ru")
```

### Квантизация перевода с русского на английский

> NameOfShadow

В данном примере наша исходная модель будет динамически квантизирована в int8

```python
from onnxruntime.quantization import quantize_dynamic, QuantType
import os

# Путь к папке с ONNX-моделями
model_dir = "onnx_models/ru-en"

# Компоненты модели MarianMT
components = [
    "encoder_model.onnx",
    "decoder_model.onnx",
    "decoder_with_past_model.onnx"
]

for component in components:
    input_path = os.path.join(model_dir, component)
    output_path = input_path.replace(".onnx", "_quant.onnx")

    # Динамическое квантование
    quantize_dynamic(
        input_path,
        output_path,
        weight_type=QuantType.QInt8,
        nodes_to_exclude=["LayerNormalization", "Gelu"],
        extra_options={"MatMulConstBOnly": True}
    )

    print(f"[+] {component} квантован -> {os.path.basename(output_path)}")

# Проверка размера файлов
print("\nРазмеры файлов после квантования:")
for component in components:
    original = os.path.getsize(os.path.join(model_dir, component)) / 1024**2
    quantized = os.path.getsize(os.path.join(model_dir, component.replace(".onnx", "_quant.onnx"))) / 1024**2
    print(f"{component}: {original:.1f} MB → {quantized:.1f} MB")
```

```python
from optimum.onnxruntime import ORTModelForSeq2SeqLM
from transformers import MarianTokenizer

# Загрузка квантованной модели
quantized_model = ORTModelForSeq2SeqLM.from_pretrained(
    "onnx_models/ru-en",
    file_name="encoder_model_quant.onnx",  # Имя квантованного файла
    provider="CPUExecutionProvider"
)

# Загрузка токенизатора
tokenizer = MarianTokenizer.from_pretrained("onnx_models/ru-en")

print("Модель и токенизатор загружены успешно!")
```

```python
input_text = "Модель успешно перевела текст!"
inputs = tokenizer(input_text, return_tensors="pt")

outputs = quantized_model.generate(**inputs)
translation = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(f"Перевод: {translation}")
```
